### 1.1 Preprocessing Steps

So, this is the step where I cleaned up the raw data to make it usable. I started by loading the **final_posts.json** and merging it with the identification and emotion labels so everything was in one main dataframe. I think it's easier to manage this way.

The raw text from social media is usually pretty messy, so I wrote a simple_clean function using regex. I have tried to over-clean it (like removing stopwords), and the result got worse, because BERT actually needs some of that context, but I did strip out URLs and fixed the weird spacing issues.

Finally, since the data comes in one file, so I have to handle the Data Splitting. I separate the original train vs. test split provided in the files, but I also carved out a 20% validation set from the training data to use as validation.

### 1.2 Feature Engineering Steps

I used the BERT Tokenizer (bert-base-uncased) to convert the cleaned sentences into input IDs.

I also had to handle the Label Encoding. Since the model cant predict strings like "anger" or "joy," I mapped each emotion to a unique integer (0, 1, 2, etc.)

Then I construct the
- **The Dataset Class:** This handles how we grab a single data point. It returns the tokenized text and the label.

- The **Collate Function:** This is really important for batching. Since not every tweet is the same length, this function dynamically pads the shorter sentences in a batch with zeros so they all match the length of the longest sentence. At first If I didn't do this, then PyTorch threw errors when trying to stack the tensors.

### 1.3 Explanation of Your Model

Instead of just using the default BERT classifier, I manually grabbed the last hidden states from BERT and applied Mean Pooling. Basically, I took the average of all the token embeddings in the sentence (ignoring the padding) to get a really robust summary vector of the whole sentence. Then, I passed that vector through a Dropout layer (to prevent overfitting) and a Linear layer to get our final predictions.

For the training loop, I added a few additonal steps:
- **Class Weights:** I noticed the dataset was imbalanced (joy was dominated), so I calculated weights and passed them to the CrossEntropyLoss. This tells the model to pay more attention to the rare classes, like digust.

- **Checkpointing:** Inside the loop, I set it up to track validation accuracy after every epoch. It only saves the model to best_model_split.ckpt if the accuracy beats the previous best score. This ensures that, I'm testing with the absolute best version of the model.

# **Competition Code**

## 1. Preprocessing Steps

```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import train_test_split

# --------------- Load Data --------------- #
df_from_file = pd.read_json('/content/final_posts.json')
emotion = pd.read_csv('/content/emotion.csv')
data_iden = pd.read_csv('/content/data_identification.csv')

# ------- Extract data from multiple files and merge them together ------- #
posts_list = [item['_source']['post'] for item in df_from_file['root'].tolist()]
df = pd.json_normalize(posts_list)

df = pd.merge(df, data_iden, left_on='post_id', right_on='id', how='left')
df = df.drop(columns=['id'])

# ------- Split train and test data --------- #
df_train = df[df['split'] == 'train']
df_train = df_train.drop(columns=['split'])
df_train = pd.merge(df_train, emotion, left_on='post_id', right_on='id', how='left')
df_train = df_train.drop(columns=['id'])

df_test = df[df['split'] == 'test']
df_test = df_test.drop(columns=['split'])

# ------------ Create Validation Split (80% Train, 20% Val) ------------ #
df_train, df_val = train_test_split(
    df_train,
    test_size=0.2,
    random_state=42,
    stratify=df_train['emotion']
)

# -------------- Text Cleaning Function --------------- #
def simple_clean(text):
    # Remove URLs if any
    text = re.sub(r'http\S+', '', text)
    # Remove extra whitespace if any
    text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply cleaning
df_train['text'] = df_train['text'].apply(simple_clean)
df_test['text'] = df_test['text'].apply(simple_clean)
df_val['text'] = df_val['text'].apply(simple_clean) # Ensure val is cleaned too
```

## 2. Feature Engineering Steps

```python
### Add the code related to the feature engineering steps in cells inside this section
import torch
from transformers import BertTokenizer
from torch.utils.data import Dataset, DataLoader

# This is inspired from the hw3 in NLP course

# -------------- Label Encoding -------------- #
classes = sorted(df_train['emotion'].unique())
label2id = {label: idx for idx, label in enumerate(classes)}
id2label = {idx: label for label, idx in label2id.items()}

df_train['label'] = df_train['emotion'].map(label2id)
df_val['label'] = df_val['emotion'].map(label2id)

# ------------ Tokenizer Initialization ------------ #
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
device = "cuda" if torch.cuda.is_available() else "cpu"

# ------------ Construct Dataset Class ---------- # For dataloader later
class EmotionDataset(Dataset):
    def __init__(self, dataframe, tokenizer, classes, has_labels=True):
        self.dataframe = dataframe
        self.tokenizer = tokenizer
        self.classes = classes
        self.text = self.dataframe.text.to_numpy()
        self.has_labels = has_labels

        if self.has_labels:
            if 'emotion' not in self.dataframe.columns:
                raise ValueError("DataFrame must contain 'emotion' column when has_labels is True.")
            if 'label' in self.dataframe.columns:
                self.labels = self.dataframe.label.to_numpy()
            else:
                label_map = {label: i for i, label in enumerate(self.classes)}
                self.labels = [label_map[l] for l in self.dataframe.emotion.to_numpy()]
        else:
            self.labels = None

    def __len__(self):
        return len(self.text)

    def __getitem__(self, idx):
        text = self.text[idx]
        item = {'text': text}
        if self.has_labels:
            label = self.labels[idx]
            item['emotion'] = label
        return item

# ------------ Collate Function (Padding) ------------ #
def collate_fn(batch):
    texts = [data['text'] for data in batch]
    encoding = tokenizer(
        texts,
        padding=True,
        truncation=True,
        return_tensors="pt"
    )

    if 'emotion' in batch[0]:
        labels = [data['emotion'] for data in batch]
        encoding['emotion'] = torch.tensor(labels, dtype=torch.long)

    encoding = {k: encoding[k].to(device) for k in encoding}
    return encoding

# ------------- DataLoaders Setup for training ------------- #
train_batch_size = 32
dl_train = DataLoader(EmotionDataset(df_train, tokenizer, classes, has_labels=True),
                      batch_size=train_batch_size, shuffle=True, collate_fn=collate_fn)
dl_val = DataLoader(EmotionDataset(df_val, tokenizer, classes, has_labels=True),
                    batch_size=1, shuffle=True, collate_fn=collate_fn)
dl_test = DataLoader(EmotionDataset(df_test, tokenizer, classes, has_labels=False),
                     batch_size=1, shuffle=False, collate_fn=collate_fn)

# df_train.emotion.value_counts() / df_train.shape[0] *100 # check the number in each emotion class
```

## 3. Model Implementation Steps

```python
### Add the code related to the model implementation steps in cells inside this section
import torch
from torch import nn
from torch.optim import AdamW
from transformers import BertModel, get_linear_schedule_with_warmup
from sklearn.utils.class_weight import compute_class_weight
from tqdm import tqdm

# ---------------- Model Architecture ---------------- #
class main_model(torch.nn.Module):
    def __init__(self, num_classes):
        super().__init__()
        self.bert = BertModel.from_pretrained("google-bert/bert-base-uncased")
        self.dropout = torch.nn.Dropout(p=0.3) # try 0.4, 0.5
        self.classification_head = torch.nn.Linear(768, num_classes)

    def forward(self, **kwargs): #forward func for prediciton
        outputs = self.bert(
            input_ids=kwargs['input_ids'],
            attention_mask=kwargs['attention_mask']
        )
        last_hidden_state = outputs.last_hidden_state
        input_mask_expanded = kwargs['attention_mask'].unsqueeze(-1).expand(last_hidden_state.size()).float()
        sum_embeddings = torch.sum(last_hidden_state * input_mask_expanded, 1)
        sum_mask = torch.clamp(input_mask_expanded.sum(1), min=1e-9)
        mean_pooled = sum_embeddings / sum_mask

        mean_pooled = self.dropout(mean_pooled)
        emotion = self.classification_head(mean_pooled)
        return {'emotion': emotion}

# -------------- Configuration hyperparameter ------------- #
lr = 3e-5
epochs = 10
model = main_model(num_classes=len(classes)).to(device)
optimizer = AdamW(model.parameters(), lr=lr)

num_training_steps = len(dl_train) * epochs
num_warmup_steps = int(0.1 * num_training_steps)
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=num_warmup_steps, num_training_steps=num_training_steps) #setup scheduler

# ------------ Compute Class Weights for Imbalanced Data ------------ #
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.array(list(label2id.values())),
    y=df_train['label'].values
)
class_weights = torch.tensor(class_weights, dtype=torch.float32).to(device)
classification_loss = torch.nn.CrossEntropyLoss(weight=class_weights)

# ------------- Training Loop ------------- #
best_val_accuracy = 0
model_save_path = 'best_model_split.ckpt'

for ep in range(epochs):
    pbar = tqdm(dl_train)
    pbar.set_description(f"Training epoch [{ep+1}/{epochs}]")
    model.train()

    for batch in pbar:
        optimizer.zero_grad()
        outputs = model(
            input_ids=batch['input_ids'],
            attention_mask=batch['attention_mask']
        )
        loss = classification_loss(outputs['emotion'], batch['emotion'])
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        scheduler.step()

    #Val Step
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for batch in dl_val:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            preds = torch.argmax(outputs['emotion'], dim=1)
            labels = batch['emotion'].to(device)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    val_acc = correct / total
    print(f"Epoch {ep+1} Validation Accuracy: {val_acc:.4f}")
    # save the model's state only the val accuracy is higher
    if val_acc > best_val_accuracy:
        best_val_accuracy = val_acc
        torch.save(model.state_dict(), model_save_path)
        print("New best model saved!")

# ---------- Prediction --------- #
model.load_state_dict(torch.load(model_save_path, weights_only=True))
model.eval()

all_preds_class = []
pbar_test = tqdm(dl_test, desc="Test")

with torch.no_grad():
    for batch in pbar_test:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        preds_class = torch.argmax(outputs['emotion'], dim=1)
        all_preds_class.extend(preds_class.cpu().tolist())

#Create Submission file
final_df = pd.DataFrame({
    'id': df_test['post_id'].reset_index(drop=True),
    'emotion': [classes[pred_id] for pred_id in all_preds_class]
})
final_df.to_csv('final_df_.csv', index=False)
```
